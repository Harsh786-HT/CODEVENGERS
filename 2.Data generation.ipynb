```python
np.random.seed(42)
rows = 1000
data = []

for _ in range(rows):
    industry_id = f"IND_{np.random.randint(1, 21)}"
    energy_kwh = np.random.randint(800, 5000)
    operating_hours = np.random.randint(8, 17)
    idle_percentage = round(np.random.uniform(5, 35), 2)
    avg_temperature = round(60 + (energy_kwh / 5000) * 180 + np.random.uniform(-5, 5), 2)
    machine_noise_db = round(45 + (energy_kwh / 5000) * 50 + np.random.uniform(-3, 3), 2)
    base_production = energy_kwh * np.random.uniform(0.18, 0.28)
    efficiency_penalty = 1 - (idle_percentage / 100)
    production_units = int(base_production * efficiency_penalty)
    fuel_type = np.random.choice(["Electricity", "Coal"])
    emission_factor = 0.82 if fuel_type == "Electricity" else 2.42
    emission_kg_co2 = round(energy_kwh * emission_factor, 2)

    data.append([
        industry_id, energy_kwh, operating_hours, idle_percentage,
        avg_temperature, machine_noise_db, production_units,
        fuel_type, emission_kg_co2
    ])

columns = [
    "industry_id", "energy_kwh", "operating_hours", "idle_percentage",
    "avg_temperature", "machine_noise_db", "production_units",
    "fuel_type", "emission_kg_co2"
]

df = pd.DataFrame(data, columns=columns)
```
